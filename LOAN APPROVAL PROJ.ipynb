# STEP 4: Exploratory Data Analysis (EDA)

# Income Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['ApplicantIncome'].hist()
plt.title("Income Distribution")
plt.show()
```

Data is right skewed

# Loan Amount

```python
sns.boxplot(x=df['LoanAmount'])
plt.show()
```

# Categorical Data

```python
df['Property_Area'].value_counts().plot(kind='bar')
plt.show()
```

# Correlation Heatmap

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
 --- LoanAmount related to income
```

# Stacked Bar Chart

```python
pd.crosstab(df['Gender'], df['Property_Area']).plot(kind='bar', stacked=True)
plt.show()
```

```python
Distribution varies across categories
```

# Key Insights:

 Applicant income and loan amount are positively related.
2. Most applicants belong to the middle-income group.
3. Semiurban areas have the highest number of applicants.
4. Credit history is mostly positive among applicants.
5. Dataset contains outliers in income and loan amount.
